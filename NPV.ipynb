```python
from ipywidgets import interact
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import HBox



@interact(EI_price=(0,200),capacity=(10,400),utilisation=(0,100),cooling =(0,100),maintenance=(0,100),disc=(0,20),
          remuneration=(0,100),EI_consumption_switch=(0,100),model=['GTX_1060_GB', 'GTX_1070','GTX_1070_ti',
                                                                    'GTX_1080','GTX_1080_ti','GTX_Titan_X',
                                                                    'GTX_Titan_Xp','Titan_V'])

       
def NPV_Calc(EI_price,capacity,utilisation,cooling,maintenance,disc,remuneration,EI_consumption_switch,model):
    
    GTX_1060_GB  ={'power': '120','cost':'299'}
    GTX_1070     ={'power': '150','cost':'449'}
    GTX_1070_ti  ={'power': '180','cost':'449'}
    GTX_1080     ={'power': '180','cost':'549'}
    GTX_1080_ti  ={'power': '250','cost':'699'}
    GTX_Titan_X  ={'power': '250','cost':'1200'}
    GTX_Titan_Xp ={'power': '250','cost':'1200'}
    Titan_V      ={'power': '250','cost':'2999'}

    if model == 'GTX_1060_GB':
        power = int(GTX_1060_GB['power'])
        cost  = int(GTX_1060_GB['cost'])
    elif model =='GTX_1070':
        power = int(GTX_1070['power'])
        cost  = int(GTX_1070['cost'])
    elif model =='GTX_1070_ti':
        power = int(GTX_1070_ti['power'])
        cost  = int(GTX_1070_ti['cost'])
    elif model == 'GTX_1080':
        power = int(GTX_1080['power'])
        cost  = int(GTX_1080['cost'])
    elif model == 'GTX_1080_ti':
        power = int(GTX_1080_ti['power'])
        cost  = int(GTX_1080_ti['cost'])
    elif model == 'GTX_Titan_X':
        power = int(GTX_Titan_X['power'])
        cost  = int(GTX_Titan_X['cost'])
    elif model == 'GTX_Titan_Xp':
        power = int(GTX_Titan_Xp['power'])
        cost  = int(GTX_Titan_Xp['cost'])
    elif model == 'Titan_V':
        power = int(Titan_V['power'])
        cost  = int(Titan_V['cost'])
    else :
        print('choose something')
    
   
        
    
    Equivalent_cap = capacity * utilisation
    
    cooling=cooling/100
    #power =180
    utilisation = utilisation/100
    remuneration = 6
    Price=549
    EI_price = 60
    EI_consumption_switch = EI_consumption_switch/100
    maintenance = maintenance/100
    disc = disc/100
    
    Unit = np.ceil(capacity*10**3/(power*(1+cooling)))
    container = np.ceil(Unit/624)
    operation_day = utilisation*365
    operation_total = np.ceil(operation_day*Unit)
    EI_consumption = capacity*8760*utilisation
    revenue = operation_total*remuneration
    capex_GPU = -Unit*cost*0.8
    capex_infra = -container* 17543
    capex_total = np.add(capex_GPU,capex_infra)
    EI_cons = -EI_consumption*EI_price/1000*EI_consumption_switch
    maintenance_price = capex_GPU*maintenance
    opex = np.add(EI_cons, maintenance_price)
    
    
    NPV_yr2 = np.ceil(np.npv(disc,[capex_total,np.add(revenue,opex),np.add(revenue,opex)]))
    NPV_yr3 = np.ceil(np.npv(disc,[capex_total,np.add(revenue,opex),np.add(revenue,opex),np.add(revenue,opex)]))
    NPV_yr4 = np.ceil(np.npv(disc,[capex_total,np.add(revenue,opex),np.add(revenue,opex),np.add(revenue,opex),
                                   np.add(revenue,opex)]))
    
    cashflow=[np.add(revenue,opex)] * 10
    
    sum_cashflowall = []
    sum_cashflow=0
    for i in cashflow:
        
        sum_cashflow += i 
        sum_cashflow = sum_cashflow
        sum_cashflowall.append(sum_cashflow)
    
    NPV_list =[]
    years = range(10)
    for year in years:   
        
        NPV = sum_cashflowall[year] / ((1 + disc)**year)
        NPV_list.append(np.ceil(NPV))
        

    print(Unit,container,operation_day,operation_total, EI_consumption,revenue,capex_GPU,capex_infra,capex_total,NPV_yr2,NPV_yr3)
         
    #print(NPV_list)
    
    plt.figure(figsize=(5,5))
    plt.plot(years,np.divide(NPV_list,1e6),color = [0,0,0],lw=2)
    plt.title("NPV vs. year")
    plt.xlabel("years")
    plt.ylabel("NPV milion $")
    
    plt.show()
    
    print(f'The total capex is {capex_total:0.0f} USD')
    print(f'The NPV for 2 years is {NPV_yr2:0.0f} USD')
    print(f'The NPV for 3 years is {NPV_yr3:0.0f} USD')
    print(f'The NPV for 4 years is {NPV_yr4:0.0f} USD')
          
    return



```
