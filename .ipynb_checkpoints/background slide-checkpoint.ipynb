```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

absolute = pd.read_table('Absolute.csv',delimiter =',')

differential = pd.read_table('Differential.csv',delimiter =',')

labels = pd.read_table('Labels.csv',delimiter =',')
```

```python
labels
```

```python

plt.plot(absolute["Date"],absolute["Tapis"])
plt.show()
```

```python
import matplotlib as mpl
import datetime

plt.rcParams['figure.dpi'] = 100

absolute=absolute.replace('-', np.nan)
absolute["Date"] = pd.to_datetime(absolute.Date) 



absolute

differential=differential.replace('-', np.nan)
differential["Date"] = pd.to_datetime(differential.Date) 


differential
```

```python
absolute[['year','month','day']] = absolute.Date.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))

differential[['year','month','day']] = differential.Date.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
```

```python
differential
```

```python
y = absolute.columns[1:-4]
col="Sakhalin Blend"
plt.rcParams['figure.dpi'] = 150
plt.rc('legend',**{'fontsize':8})


for col in y:
    plot_data = absolute[col].dropna()
    K = list(plot_data.index.values) 
    plt.plot(absolute["Date"].iloc[K],plot_data.astype(float),'--',label=col)
    BD = plt.plot(absolute["Date"],absolute["Dtd Brent"],'k')
    plt.xlabel("Date")
    plt.ylabel("Absolute value")
    plt.xlim(['2008-10-01','2020-01-01'])
    #plt.legend()
    plt.legend((BD),('Dated Brent',),loc='upper right') 
```
