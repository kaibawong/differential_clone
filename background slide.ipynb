```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  

absolute = pd.read_table('Absolute.csv',delimiter =',')
```

```python
absolute

plt.plot(absolute["Date"],absolute["Tapis"])
plt.show()
```

```python
import matplotlib as mpl
import datetime

plt.rcParams['figure.dpi'] = 100


absolute["Date"] = pd.to_datetime(absolute.Date) 

absolute=absolute.replace('-', np.nan)

absolute
```

```python
absolute[['year','month','day']] = absolute.Date.apply(lambda x: pd.Series(x.strftime("%Y,%m,%d").split(",")))
```

```python
absolute
```

```python
y = absolute.columns[1:-4]
col="Sakhalin Blend"
plt.rcParams['figure.dpi'] = 150

for col in y:
    plot_data = absolute[col].dropna()
    K = list(plot_data.index.values) 
    plt.plot(absolute["Date"].iloc[K],plot_data.astype(float),'--',label=col)
    plt.plot(absolute["Date"],absolute["Dtd Brent"],'k')
    plt.xlabel("Date")
    plt.ylabel("Absolute value")
    plt.legend()
      
    
```
